```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df_tips = sns.load_dataset("tips")

Q1 = df_tips['total_bill'].quantile(0.25)
Q3 = df_tips['total_bill'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df_tips[(df_tips['total_bill'] < lower_bound) | (df_tips['total_bill'] > upper_bound)]

print("--- Outlier Analysis Results ---")
print(f"Number of observations: {len(df_tips)}")
print(f"Q1 (25th percentile): ${Q1:.2f}")
print(f"Q3 (75th percentile): ${Q3:.2f}")
print(f"IQR: ${IQR:.2f}")
print(f"Lower Bound: ${lower_bound:.2f}")
print(f"Upper Bound: ${upper_bound:.2f}")
print(f"Number of outliers detected: {len(outliers)}")
print("\nIdentified Outliers:")
print(outliers[['total_bill', 'tip', 'sex', 'day']])

plt.figure(figsize=(8, 6))
sns.boxplot(y=df_tips['total_bill'], color='magenta')
plt.title('Boxplot of Total Bill Amount (with outliers)')
plt.ylabel('Total Bill ($)')
plt.show()
```
