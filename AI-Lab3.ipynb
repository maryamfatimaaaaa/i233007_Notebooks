Task 2

```python
agent_model = {
    "A": "Unknown",
    "B": "Unknown"
}

environment = {
    "A": "Dirty",
    "B": "Dirty"
}

agent_location = "A"


def get_percept(env, location):
    return location, env[location]


def update_model(model, percept):
    room, status = percept
    model[room] = status


def decide_action(model, location):
    if model[location] == "Dirty":
        return "Clean"
    elif location == "A":
        return "MoveRight"
    elif location == "B":
        return "MoveLeft"
    else:
        return "NoOp"


def execute_action(env, action, location):
    if action == "Clean":
        env[location] = "Clean"
    elif action == "MoveRight":
        location = "B"
    elif action == "MoveLeft":
        location = "A"
    return location


steps = 6
for step in range(1, steps + 1):
    percept = get_percept(environment, agent_location)
    update_model(agent_model, percept)
    action = decide_action(agent_model, agent_location)
    agent_location = execute_action(environment, action, agent_location)

    print(f"Step {step}")
    print(f"Percept: {percept}")
    print(f"Action: {action}")
    print(f"Environment: {environment}")
    print(f"Agent Model: {agent_model}")
```

Task 3

```python
routes = {
    "Route A": {"distance": 10, "time": 15, "risk": "Low"},
    "Route B": {"distance": 8,  "time": 20, "risk": "Medium"},
    "Route C": {"distance": 12, "time": 10, "risk": "High"}
}

risk_values = {
    "Low": 1,
    "Medium": 2,
    "High": 3
}


def utility(distance, time, risk):
    return -(distance + time + risk)

utility_scores = {}

for route, attributes in routes.items():
    distance = attributes["distance"]
    time = attributes["time"]
    risk = risk_values[attributes["risk"]]
    utility_scores[route] = utility(distance, time, risk)

best_route = max(utility_scores, key=utility_scores.get)

print("Utility Scores:")
for route, score in utility_scores.items():
    print(f"{route}: {score}")

print("\nSelected Route:", best_route)
```
