Lab Task 1

```python
num=[2, 1, 6, 5]
def is_even(num):
    return num%2==0
en=filter(is_even, num)
squared_list = [n ** 2 for n in en]
print(squared_list)
```

```python
def is_odd(num):
    return num%2!=0
od=filter(is_odd, num)
cubed_list = [n ** 2 for n in od]
print(cubed_list)
```

```python
merged=squared_list+cubed_list
print(sorted(merged))
```

```python
import numpy as np


performance_scores = np.array([
    [85, 90, 78],  
    [92, 88, 95],  
    [70, 75, 80],  
    [95, 98, 92]   
])

print("Original Scores:\n", performance_scores, "\n")

average_scores = np.mean(performance_scores, axis=1)
print("Average Scores per System:", average_scores, "\n")

sorted_indices_asc = np.argsort(average_scores)
print("Ascending Sorted Indices:", sorted_indices_asc, "\n")
sorted_indices_desc = sorted_indices_asc[::-1]

print("--- System Ranking (Best to Worst) ---")
for i in sorted_indices_desc:
    print(f"System {chr(65+i)} (Avg: {average_scores[i]:.2f}): {performance_scores[i]}")

print("\n--- Ranked Average Scores ---")
ranked_averages = average_scores[sorted_indices_desc]
for rank, avg_score in enumerate(ranked_averages):
    print(f"Rank {rank+1}: {avg_score:.2f}")
```
